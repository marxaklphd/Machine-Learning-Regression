```python
'''Elastic Net Regression'''
from sklearn.linear_model import ElasticNet
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
import numpy as np
pd.options.mode.chained_assignment = None
import warnings
warnings.filterwarnings('ignore')
import statsmodels.formula.api as smf
from sklearn.preprocessing import add_dummy_feature
```

```python
m=100
x = 4 * np.random.rand(m,1)
y = 1 + (0.4 * x) + np.random.randn(m,1) / 2
plt.plot(x,y,'.')
```

```python
en_regression = ElasticNet(alpha=0.1,l1_ratio=0.5)
en_regression.fit(x,y)
```

```python
x_topredict = np.linspace(0,4,100).reshape(100,1)
y_predicted = en_regression.predict(x_topredict)
plt.plot(x,y,'b.',label='Data')
plt.plot(x_topredict,y_predicted,'r--',label='Predicted')
plt.legend()
plt.xlabel("$x$")
plt.ylabel("$y$  ", rotation=0)    
plt.grid()
plt.show();
```
